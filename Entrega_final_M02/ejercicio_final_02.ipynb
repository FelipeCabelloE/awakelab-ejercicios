```python
# Importamos los paquetes
from IPython.display import display
import pandas as pd
```

```python
state_abbrevs_df = pd.read_csv('state-abbrevs.csv' ,delimiter=";")
state_area_df = pd.read_csv('state-area.csv',delimiter=";")
state_population_df = pd.read_csv('state-population.csv',delimiter=";")
# state_population_df = state_population_df.dropna()
# state_population_df = state_population_df.rename(columns={'State': 'state'})
```

```python
dataframe_list = [state_abbrevs_df, state_area_df, state_population_df]
```

```python
# Exploración inicial
for df in dataframe_list:
    display(df.head())
```

```python
# Elimianmos columnas vacías
dataframe_list_dropna = [df.dropna(axis='columns', how='all') for df in dataframe_list]


for df in dataframe_list_dropna:
    display(df.head())
```

```python



#group by state/region and ages for min max count and mean
display(dataframe_list_dropna[2].groupby(['state/region','ages']).mean())
display(dataframe_list_dropna[2].groupby(['state/region','ages']).count())
display(dataframe_list_dropna[2].groupby(['state/region','ages']).min())
display(dataframe_list_dropna[2].groupby(['state/region','ages']).max())
# list the grouped df
```

```python
#5. Forme un nuevo dataset llamado df que tenga la información de los 3 sets importados. Use las llaves necesarias.

dataframe_list_dropna[2] = dataframe_list_dropna[2].rename(columns={'state/region': 'abbreviation'})

df = dataframe_list_dropna[0].merge(dataframe_list_dropna[2], on='abbreviation', how='left').merge(dataframe_list_dropna[1], on='state', how='left')

df = df[df['ages'] != 'under18']

```

```python
df_mean = df.groupby(['state','abbreviation', 'ages']).mean()
```

```python
df_mean['pop/area'] = df_mean['population']/df_mean['area (sq. mi)']
```

```python
df_mean.head()
```

```python
# 7. cree un dataframe que posea: El estado, el área, la población, la población por área y como índice la abreviación de nombre del estado.




final_df = dataframe_list_dropna[0]
final_df.set_index('abbreviation', inplace=True)
final_df['poblacion'] = df_mean['population'].values
final_df['area'] = df_mean['area (sq. mi)'].values
final_df['poblacion_por_area'] = df_mean['pop/area'].values

final_df.head()
```
