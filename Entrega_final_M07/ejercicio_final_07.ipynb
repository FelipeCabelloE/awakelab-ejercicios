```python

import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Input, Dense
from keras.utils import to_categorical
from keras.models import Model
import numpy as np
from sklearn.preprocessing import StandardScaler

import warnings
warnings.filterwarnings('ignore')
```

```python
# 2. Importe la base de datos pima-indians-diabetes.csv
df = pd.read_csv('pima-indians-diabetes.csv', header= None)

#set column names
df.columns = ['preg', 'plas', 'pres', 'skin', 'test', 'mass', 'pedi', 'age', 'class']

df.head(5)
```

```python
df.describe()
```

```python
df.info()
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(2, activation='softmax'))
# set X, y
X = df.drop('class', axis=1)
y = df['class']

# split data
y = to_categorical(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=123)

# scale data
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# compile model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

#show model summary
model.summary()
```

```python
# fit model
model.fit(X_train, y_train, epochs=150, batch_size=10, verbose=1)
```

```python
# evaluate model
scores = model.evaluate(X_test, y_test, verbose=1)
print("Accuracy: %.2f%%" % (scores[1]*100))

# con un 74 % de precision, el modelo podría ser mejorardo a travez de un mayor preprocesamiento de la informacion, normalizando y densificando la red neuronal

```
