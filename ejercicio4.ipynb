2. Cree una clase que nombrará Cl i e nt e con los siguientes atributos:
persona (puede ser el titular de una cuenta) y monto (puede tener
decimales). El atributo persona será obligatorio y el monto es
opcional. Debe considerar lo siguiente para la clase pedida:
● Un constructor, donde los datos pueden estar vacíos.

● Los setters y getters para cada uno de los atributos. El
atributo no se puede modificar directamente, sólo ingresando
o retirando dinero.
● revelar() : Revela la información del Cliente.
● ingresar(monto) : Registra el monto del cliente para una
cuenta, si el número que ingresa es negativo, no hará nada.
● retirar(monto) : Registra un posible retiro de la cuenta del
cliente. El cliente puede tener un saldo negativo.

```python
from dataclasses import dataclass, field

@dataclass()
class Cliente():
    _persona: str
    monto: float = field(default_factory=float)

    @property
    def persona(self):
        return self._persona

    @persona.setter
    def persona(self, value):
        self._persona = value

    def retirarMonto(self, monto):
        if monto <= 0:
            raise ValueError("El monto debe ser mayor que 0")
        else:
            self.monto -= monto

    def agregarMonto(self, monto):
        if monto <= 0:
            raise ValueError("El monto debe ser mayor que 0")
        self.monto += monto

    @property
    def getMonto(self):
        return self.monto

    def revelar(self):
        return f"Cliente: {self._persona} Monto: {self.monto}"

    
    
cuenta_corriente = Cliente("Juan")
cuenta_corriente.retirarMonto(100)
print(cuenta_corriente.revelar())
cuenta_corriente.agregarMonto(100)
print(cuenta_corriente.revelar())
cuenta_corriente.agregarMonto(100)
cuenta_corriente.agregarMonto(100)
print(cuenta_corriente.revelar())
```
