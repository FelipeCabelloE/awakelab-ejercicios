```python
import pandas as pd

```

```python
# Se genera el dataframe
df_titanic = pd.read_csv('train.csv', index_col="PassengerId")
```

```python
# Dimensiones, columnas, tipos de datos, primeras 10 filas y Últimas 10 filas

info_dataframe = [df_titanic.shape, df_titanic.columns, df_titanic.dtypes, df_titanic.head(10), df_titanic.tail(10)]

for info in info_dataframe:
    print(info)
```

```python
# Mostrar pasajero 148

df_titanic.loc[148]
```

```python
# Mostrar por pantalla las filas pares del dataframe df_titanic
df_titanic.loc[df_titanic.index % 2 == 0]
```

```python
# Mostrar nombres de las personas que iban en primera clase ordenadas alfabeticamente

is_firstclass = df_titanic['Pclass'] == 1

df_titanic[is_firstclass].sort_values('Name')

```

```python

# porcentaje de personas muertas y vivas

porcentaje_survived = df_titanic.Survived.value_counts(normalize=True)

print(f"Murió un  {porcentaje_survived[0]*100:.2f}% de la gente y vivió un {porcentaje_survived[1]*100:.2f}% de la gente")
```

```python
# porcentaje muertos por clase


for value in df_titanic.Pclass.unique():
    mascara_clase = df_titanic.Pclass == value
    mascara_vivos = df_titanic.Survived == 1
    porcentaje_vivos_clase = df_titanic[mascara_clase & mascara_vivos].Survived.count() / df_titanic[mascara_clase].Survived.count() * 100
    print(f"Porcentaje de vivos en clase {value} es {porcentaje_vivos_clase:.2f}%")
    
    

```

```python
# elimina las filas con valores na en la columan Age

df_titanic_dropnaAge = df_titanic.dropna(subset=['Age'])
```

```python
# mask females

mascara_mujer = df_titanic_dropnaAge['Sex'] == 'female'

df_titanic_dropnaAge.loc[mascara_mujer, 'Survived'].count() / df_titanic_dropnaAge.Survived.count()
```

```python
for value in df_titanic.Pclass.unique():
    mascara_clase = df_titanic.Pclass == value
    porcentaje_mujer_clase = df_titanic_dropnaAge.loc[mascara_clase, "Sex"].value_counts(normalize=True)[1] * 100
    print(f"Porcentaje de mujeres en clase {value} es {porcentaje_mujer_clase:.2f}%")
    
```
