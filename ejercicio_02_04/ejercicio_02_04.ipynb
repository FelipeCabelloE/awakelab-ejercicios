```python

# porcentaje de personas muertas y vivas

porcentaje_survived = df_titanic.Survived.value_counts(normalize=True)

print(f"MuriÃ³ un  {porcentaje_survived[0]*100:.2f}% de la gente y viviÃ³ un {porcentaje_survived[1]*100:.2f}% de la gente")
```

```python
# porcentaje muertos por clase


for value in df_titanic.Pclass.unique():
    mascara_clase = df_titanic.Pclass == value
    mascara_vivos = df_titanic.Survived == 1
    porcentaje_vivos_clase = df_titanic[mascara_clase & mascara_vivos].Survived.count() / df_titanic[mascara_clase].Survived.count() * 100
    print(f"Porcentaje de vivos en clase {value} es {porcentaje_vivos_clase:.2f}%")
    
    

```

```python
# elimina las filas con valores na en la columan Age

df_titanic_dropnaAge = df_titanic.dropna(subset=['Age'])
```

```python
# mask females

mascara_mujer = df_titanic_dropnaAge['Sex'] == 'female'

df_titanic_dropnaAge.loc[mascara_mujer, 'Survived'].count() / df_titanic_dropnaAge.Survived.count()
```

```python


# Calcular la edad media de las mujeres en cada clase
avg_age_female_by_class = df_titanic_dropnaAge[df_titanic_dropnaAge['Sex'] == 'female'].groupby('Pclass')['Age'].mean()

# Mostrar por pantalla la edad media de las mujeres en cada clase
print(avg_age_female_by_class)
```

```python
# AÃ±adir una nueva columna 'IsMinor' basada en la edad del pasajero
df_titanic_dropnaAge['IsMinor'] = df_titanic_dropnaAge['Age'] < 18

# Mostrar las primeras filas del dataset con la nueva columna 'IsMinor'
print(df_titanic_dropnaAge.head())
```

```python
# Calcular el porcentaje de menores y mayores de edad que sobrevivieron en cada clase
survival_percentage_by_age_class = df_titanic_dropnaAge.groupby(['Pclass', 'IsMinor'])['Survived'].mean() * 100

# Mostrar por pantalla el porcentaje de supervivencia por edad y clase
print(survival_percentage_by_age_class)
```
