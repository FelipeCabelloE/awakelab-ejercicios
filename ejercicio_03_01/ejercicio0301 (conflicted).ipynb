```python
import pandas as pd
import matplotlib as plt

# lista de niños 2 6 10 5 10 3 2 2

niños = [2, 6, 10, 5, 10, 3, 2, 2]

df_niños = pd.DataFrame({"metros": niños})
```

```python
# estadistica descriptiva de df_niños

df_niños.describe()
```

```python
## frecuencia absoluta

df_niños.metros.plot(kind='bar', figsize=(10,5), title='Distribución de metros por niño')
```

```python
# grafico de barras para las frecuencias acumuladas para el dataframe df_niños


df_niños.metros.cumsum().plot(kind='bar', title='Frecuencia acumulada')



```

```python
# frecuencia relativa


total = df_niños.metros.sum()

freq_relativa = df_niños.metros / total

freq_relativa.plot(kind='bar')
```

```python
# valores entre los que se encuentra el 75% de los datos

tercer_cuartil = df_niños.metros.quantile(0.75)
primer_cuartil = df_niños.metros.quantile(0)

print(f"entre el valor {primer_cuartil} y {tercer_cuartil} se puede encontrar el 75% de los datos")
```

```python
# moda

df_niños.metros.mode()

# varianza

df_niños.metros.var()

```
