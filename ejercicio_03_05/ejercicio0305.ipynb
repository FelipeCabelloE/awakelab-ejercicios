Ahora con los datos anteriores construya tablas pivote según las siguientes
instrucciones:
1. Construya una tabla fijando “continent”, con las funciones vistas en
clase
2. Realice una tabla de doble entrada, fijando “continent” y “year”.
Comente sus resultados.
3. Con las variables anteriores, obtenga el promedio de la variable “pop”.
4. Obtenga un gráfico de barras del punto 3, comente sus resultados.

```python
import numpy as np
gapminder.pivot_table(index='continent', columns='year', values=['lifeExp', 'pop'], aggfunc=np.mean)
```

```python
gapminder.pivot_table(index=['continent', 'year'], values=['lifeExp', 'pop'], aggfunc=np.mean)
```

```python
gapminderPivotContYear_Pop = gapminder.pivot_table(index=['continent', 'year'], values=['pop'], aggfunc=np.mean)
```

```python
gapminderPivotContYear_Pop.index
```

```python
# resetear el index para poder plotear
gapminderPivotContYear_Pop_reset = gapminderPivotContYear_Pop.reset_index()

# visualizar
plt.figure(figsize=(15,8))
sns.barplot(x='year', y='pop', hue='continent', data=gapminderPivotContYear_Pop_reset)
plt.title('Población promedio por continente y año')
plt.show()
```

- La población mundial ha aumentado con los años.
- Africa superó en población a oceanía y las américas a europa, durante el siglo XX.
- Asia es el continente con más población.
