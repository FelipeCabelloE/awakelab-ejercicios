Pregunta 1:

Aprendizaje supervisado: Es un tipo de aprendizaje de máquina donde se entrena un modelo con datos etiquetados. El modelo aprende a predecir la etiqueta de un dato nuevo basándose en las características de los datos de entrenamiento.

Aprendizaje sin supervisión: Este tipo de aprendizaje de máquina utiliza datos no etiquetados. El modelo aprende a identificar patrones y estructuras en los datos sin tener una etiqueta objetivo predefinida.

Aprendizaje por refuerzo: En este tipo de aprendizaje, un agente aprende a tomar decisiones basándose en recompensas y castigos. El objetivo es maximizar la recompensa total.

Algoritmos de regresión: Son algoritmos de aprendizaje supervisado que predicen un valor continuo (como el precio de una casa) basándose en las características de los datos.

Algoritmos de agrupación: Son algoritmos de aprendizaje no supervisado que agrupan los datos en diferentes categorías basándose en su similitud.

Algoritmos de Aprendizaje Profundo: Son algoritmos que utilizan redes neuronales con muchas capas (profundas) para aprender de los datos. Son especialmente buenos para tareas complejas como el reconocimiento de imágenes y el procesamiento del lenguaje natural.
