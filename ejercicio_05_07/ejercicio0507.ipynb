La distribuciÃ³n es similar en el train y test set, ademÃ s entre las categorÃ­as tambiÃ©n

```python
# split x and y

from json import encoder
from sklearn.discriminant_analysis import StandardScaler


X_train = train.drop(['Activity', 'subject'], axis=1)
X_test = test.drop(['Activity', 'subject'], axis=1)
# encode y_train and y_test

from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()

y_train = encoder.fit_transform(train['Activity'])
y_test = encoder.fit_transform(test['Activity'])

# Initialize the standard scaler
scaler = StandardScaler()

x_train_scaled = scaler.fit_transform(X_train)
x_test_scaled = scaler.fit_transform(X_test)


```

```python
# Create a Random Forest classifier
from sklearn.ensemble import RandomForestClassifier


rf = RandomForestClassifier(
    n_estimators=1000,
    random_state=0
)

# Train the classifier
rf.fit(x_train_scaled, y_train)
```

```python
rf.score(x_test_scaled, y_test)
```

No son muy claros en quÃ© modelo o parÃ¡metros utilizar. Lo mÃ¡s directo es hacer un random forest

```python
from sklearn import tree
# Get the decision tree structure from the Random Forest
trees = rf.estimators_

# Visualize the first decision tree in the Random Forest
fig, ax = plt.subplots(figsize=(12, 8))
ax.set_axis_off()
_ = tree.plot_tree(
    trees[-1],
    feature_names=train.drop(['Activity', 'subject'], axis=1).columns,
    class_names=train['Activity'].unique(),
    filled=True
)
#save the figure high resolution
fig.savefig('rf_individualtree.png', dpi=300)
```
